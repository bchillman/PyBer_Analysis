### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os

# File to Load (Remember to change these)
city_data_to_load = os.path.join('.', 'Resources', "city_data.csv")
ride_data_to_load = os.path.join('.', 'Resources', "ride_data.csv")

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df.head()
```

## Deliverable 1: Get a Summary DataFrame 

```python
#  1. Get the total rides for each city type
pyber_rides_type = pyber_data_df.groupby('type').count()['ride_id']
pyber_rides_type
```

```python
# 2. Get the total drivers for each city type
pyber_drivers_type = pyber_data_df.groupby('type').sum()['driver_count']
pyber_drivers_type
```

```python
#  3. Get the total amount of fares for each city type
pyber_fares_type = pyber_data_df.groupby('type').sum()['fare']
pyber_fares_type
```

```python
#  4. Get the average fare per ride for each city type. 
average_fares_ride = pyber_fares_type.div(pyber_rides_type)
average_fares_ride
```

```python
# 5. Get the average fare per driver for each city type. 
average_fare_driver = pyber_fares_type.div(pyber_drivers_type)
average_fare_driver
```

```python
#  6. Create a PyBer summary DataFrame. 
summary_df = pyber_data_df.groupby('type').agg(total_rides=pd.NamedAgg('ride_id', 'count'),
                                              total_drivers=pd.NamedAgg('driver_count', 'sum'),
                                              total_fare=pd.NamedAgg('fare', 'sum'))
# Don't know how to add these columns with agg function so I am adding them seperately
summary_df['Average Fare per Ride'] = average_fares_ride
summary_df['Average Fare per Driver'] = average_fare_driver

# Rename columns to get rid of bad formatting
summary_df = summary_df.rename({'total_rides': 'Total Rides', 'total_drivers': 'Total Drivers', 'total_fare': 'Total Fare'}, 
                               axis=1)
summary_df
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
summary_df.index.name = None
summary_df
```

```python
#  8. Format the columns.
format_dict = {'Total Fare':'${:,.2f}', 'Average Fare per Ride': '${:.2f}', 'Average Fare per Driver': '${:.2f}'}
summary_df = summary_df.style.format(format_dict)
summary_df
```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# 1. Read the merged DataFrame
pyber_data_df.head()
```

```python
# 2. Using groupby() to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date.
pyber_city_date = pyber_data_df.groupby(['type', 'date']).sum()['fare']
pyber_city_date.head()
```

```python
# 3. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
# df = df.reset_index()
pyber_city_date = pyber_city_date.reset_index()
pyber_city_date.head()
```

```python
# 4. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 
pyber_pivot = pyber_city_date.pivot(index='date', columns='type',values='fare')
pyber_pivot
```

```python
# 5. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-29'.
fares_2019 = pyber_pivot.loc['2019-01-01':'2019-04-29',:]
fares_2019.tail()
```

```python
# 6. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
# df.index = pd.to_datetime(df.index)
fares_2019.index = pd.to_datetime(fares_2019.index)
fares_2019
```

```python
# 7. Check that the datatype for the index is datetime using df.info()
fares_2019.info()
```

```python
# 8. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.
fares_weekly = fares_2019.resample('W').sum()
fares_weekly
```

```python
# 8. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 
fig = plt.figure(figsize=(10,4))
ax1 = fig.add_subplot()
fares_weekly.plot(ax=ax1)
ax1.set_xlabel('Date')
ax1.set_ylabel('Fare (USD)')
ax1.set_title('Total Fare by City Type Over Time')
#ax.set_xticks(x_axis)
ax1.legend(['Rural', 'Suburban', 'Urban'])
# Import the style from Matplotlib.
from matplotlib import style
# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')

# Save the figure
plt.savefig("./Analysis/PyBer_fare_summary.png")
```
